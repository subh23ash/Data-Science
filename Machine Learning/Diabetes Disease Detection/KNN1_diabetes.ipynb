```python
#Training and testing data
from sklearn.model_selection import train_test_split #fn used to split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)#x in dataframe...... to get array we put .values
x_train
```

```python
#Normalization
# more preference to higher columns...if do not done normalization
# METHOD 1
# 1)standard scalar :works based on mathematical eqn
# eqn==>   z=(x-u)/s

# u== Mean of training data
# s== std deviation of training data
# z=new value in column
# x== old value in column
#values in aspecific range
```

```python
from sklearn.preprocessing import StandardScaler    #preprocessing
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)      #fit used for model creation
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
# #PERFORMANCE EVALUATION

# 1) ACCURACY SCORE

# 2)RECALL

# 3)PRECISION

# 4)F1_SCORE
```

```python
#performance evaluation

from sklearn.metrics import confusion_matrix
mat=confusion_matrix(y_test,y_pred)
print(mat)                                                               #tp    fp
                                                                         #fn    tn
```

```python
from sklearn.metrics import accuracy_score               #y_test==>actual value    #y_pred===>predicted value
score=accuracy_score(y_test,y_pred)
score                                                    #accuracy score for each machine is different...each time the training and testing data for
                                                         # each machine is different
```

```python
#going to predict a value
pred=knn.predict(scaler.transform([[3,130,75,35,3,26,0.16,35]]))
pred
#yes he is a diabetic patient
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
